```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd  
import numpy as np
import seaborn as sns
#from sklearn.svm import SVC
#from sklearn.model_selection import KFold
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python

data=pd.read_csv('android.csv')
```

```python
data
```

```python
data.shape
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data.head()
```

```python
import seaborn as sns
```

```python
sns.countplot(x='malware',data=data)
```

```python
target_count = data.malware.value_counts()
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
```

```python
count_class_0, count_class_1 = data.malware.value_counts()
```

```python
df_class_0 = data[data['malware'] == 0]
df_class_1 = data[data['malware'] == 1]
```

```python
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_test_over = pd.concat([df_class_0, df_class_1_over], axis=0)
```

```python
df_test_over.shape
```

```python
sns.countplot(x='malware',data=df_test_over)
```

```python
X=df_test_over.iloc[:,df_test_over.columns !='malware']
Y=df_test_over.iloc[:,df_test_over.columns =="malware"]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.utils import shuffle
```

```python
X, Y=shuffle(X, Y)
```

```python

X.head()
```

```python
X=X.drop(columns='name')
X.head()
```

```python
Y.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,Y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  
featureScores.nlargest(10,'Score')
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,Y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(X_train,y_train)
```

```python
y_pred = tree.predict(X_test)
y_pred
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))

print("Recall:",metrics.recall_score(y_test, y_pred))
```

# SVM Algorithm

```python
from sklearn import datasets
cancer = datasets.load_breast_cancer()
```

```python
print("Features: ", cancer.feature_names)
print("Labels: ", cancer.target_names)
```

```python
cancer.data.shape
```

```python
print(cancer.data[0:5])
```

```python
print(cancer.target)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
from sklearn import svm


clf = svm.SVC(kernel='linear') 
clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))

print("Recall:",metrics.recall_score(y_test, y_pred))
```
